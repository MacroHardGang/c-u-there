```python
''' Performs the sobel edge detection based on a value threshold '''
def sobel_edge_by_value(img, threshold):
    # perform sobel detect in x and y directions, then use magnitude
    sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)  
    sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)  
    magnitude = cv2.magnitude(sobelx, sobely)
    # apply threshold
    _,result = cv.threshold(img,threshold,255,cv.THRESH_TOZERO)
    return result
```
