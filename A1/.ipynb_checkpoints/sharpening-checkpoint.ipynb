## 1.2 - Sharpening
Sharpening is deﬁned as follow ->
**sharpened image = original image + details** where **details = original image − blurred image**

#### 1. Sharpen the given image ‘rice’ (see Figure 2(a)) where: 
* the blurred image is generated using 5×5 box ﬁlter. 
* the blurred image is generated using 5×5 Gaussian ﬁlter with variance = 2. 


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# load original image
original_img = cv2.imread('rice.png')

# display original image
plt.axis('off')
plt.imshow(original_img, cmap='gray')
plt.title("Original Image")
plt.show()
```

```python
# apply gaussian and box filters to create blurred images
box_filter = cv2.filter2D(original_img, -1, np.ones((5,5), np.float32)/(5*5))
gaussian_filter = cv2.GaussianBlur(original_img, (5,5), np.sqrt(2))

# shapren image using blurred images
box_sharpened = cv2.add(original_img, cv2.subtract(original_img, box_filter) )
gaussian_sharpened = cv2.add(original_img, cv2.subtract(original_img, gaussian_filter) )

# display results
plt.figure(figsize=(10,10))
plt.subplot(121), plt.imshow(box_sharpened) 
plt.title("Sharpened Image (Box Filter)"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(gaussian_sharpened)
plt.title("Sharpened Image (Gaussian Filter)"), plt.xticks([]), plt.yticks([])
plt.show()
```

#### 2. Which of the two methods is expected to give better results and why? Can you observe expected result?

We can expect that the box filter will create a sharper image. Since a box filter generates a blurrier image, it will provide more 'details' generated by the formula above. As a result, the final sharpened image can be expected to be sharper with a box filter. This can also be observed in the results above.
