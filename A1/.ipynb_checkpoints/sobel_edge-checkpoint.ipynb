## 2.1 - Sobel Edge Detector

#### 1. Eﬀect of sharpening on edge detection 
* Apply a Sobel edge detector to the image ‘rice’ (Figure 2(a)). Use following thresholds: 150 and 200. Use kernel size of 3.
* Apply a Sobel edge detector to the two previously-sharpened images with the same threshold. 
* Comment on the eﬀectiveness of using sharpening prior to the edge detection. 

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage.measure as skimg

# load original image
original_img = cv2.imread('rice.png', cv2.IMREAD_GRAYSCALE)

# display original image
plt.axis('off')
plt.imshow(original_img, cmap='gray')
plt.title("Original Image")
plt.show()
```

```python
''' Performs the sobel edge detection based on a threshold value'''
def sobel_edge_by_value(img, threshold):
    # perform sobel detect in x and y directions, then use magnitude
    sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)  
    sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)  
    magnitude = cv2.magnitude(sobelx, sobely)
    # apply threshold
    _,result = cv2.threshold(magnitude,threshold,255,cv2.THRESH_TOZERO)
    return result

# perform sobel edge detect with thresholds of 150 and 200
sobel_150 = sobel_edge_by_value(original_img, 150)
sobel_200 = sobel_edge_by_value(original_img, 200)

# display results
plt.figure(figsize=(10,10))
plt.subplot(121), plt.imshow(sobel_150, cmap='gray') 
plt.title("Sobel Edge Detection (Threshold=150)"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(sobel_200, cmap='gray') 
plt.title("Sobel Edge Detection (Threshold=200)"), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# create sharpened images
box_filter = cv2.filter2D(original_img, -1, np.ones((5,5), np.float32)/(5*5))
gaussian_filter = cv2.GaussianBlur(original_img, (5,5), np.sqrt(2))
box_sharpened = cv2.add(original_img, cv2.subtract(original_img, box_filter) )
gaussian_sharpened = cv2.add(original_img, cv2.subtract(original_img, gaussian_filter) )

# perform sobel edge detect with thresholds of 150 and 200 on each of the sharpened images
sobel_150_box = sobel_edge_by_value(box_sharpened, 150)
sobel_200_box = sobel_edge_by_value(box_sharpened, 200)
sobel_150_gaussian = sobel_edge_by_value(gaussian_sharpened, 150)
sobel_200_gaussian = sobel_edge_by_value(gaussian_sharpened, 200)

# display results
plt.figure(figsize=(10,10))
plt.subplot(121), plt.imshow(sobel_150_box, cmap='gray') 
plt.title("Sharpened With Box Filter (Threshold=150)"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(sobel_200_box, cmap='gray') 
plt.title("Sharpened With Box Filter (Threshold=200)"), plt.xticks([]), plt.yticks([])
plt.show()

plt.figure(figsize=(10,10))
plt.subplot(121), plt.imshow(sobel_150_gaussian, cmap='gray') 
plt.title("Sharpened With Gaussian Filter (Threshold=150)"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(sobel_200_gaussian, cmap='gray') 
plt.title("Sharpened With Gaussian Filter (Threshold=200)"), plt.xticks([]), plt.yticks([])
plt.show()
```

##### Prior to sharpening, a threshold of 200 was not enough to precisely detect the edges. After sharpening, a threshold of 200 was enough to precisely detect the edges. Based on the results shown above, it is shown clearly that edge detection works better when an image is sharpened before hand. Additionally, since a box filter was shown to sharpen an image better in the previous question, using a box filter also makes the edge detection better as a direct result.

#### 2. Eﬀect of denoising on edge detection 
* Apply a Sobel edge detector to the image ‘woman’ (see Figure 2(b)). Use two values of thresholds: 10% and 20% of the maximum ﬁlter response.  
* First denoise image with a 3×3 box ﬁlter and then apply a Sobel edge detector. Use the same values of threshold.  
* Comment on the eﬀectiveness of using denoising prior to edge detection.

```python
# load original image
original_img = cv2.imread('woman.png', cv2.IMREAD_GRAYSCALE)

# display original image
plt.axis('off')
plt.imshow(original_img, cmap='gray')
plt.title("Original Image")
plt.show()
```

```python
''' Performs the sobel edge detection based on a threshold ratio'''
def sobel_edge_by_ratio(img, ratio):
    # perform sobel detect in x and y directions, then use magnitude
    sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)  
    sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)  
    magnitude = cv2.magnitude(sobelx, sobely)
    # apply threshold
    _, maxVal, _, _ = cv2.minMaxLoc(magnitude)
    _,result = cv2.threshold(magnitude,ratio*maxVal,255,cv2.THRESH_TOZERO)
    return result

# perform sobel edge detect with thresholds of 10% and 20%
sobel_150 = sobel_edge_by_ratio(original_img, 0.1)
sobel_200 = sobel_edge_by_ratio(original_img, 0.2)

# display results
plt.figure(figsize=(10,10))
plt.subplot(121), plt.imshow(sobel_150, cmap='gray') 
plt.title("Sobel Edge Detection (Threshold=150)"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(sobel_200, cmap='gray') 
plt.title("Sobel Edge Detection (Threshold=200)"), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# denoise image using box filter
denoised_img = cv2.filter2D(original_img, -1, np.ones((3,3), np.float32)/(3*3))

# perform sobel edge detect with thresholds of 10% and 20%
sobel_150 = sobel_edge_by_ratio(denoised_img, 0.1)
sobel_200 = sobel_edge_by_ratio(denoised_img, 0.2)

# display results
plt.figure(figsize=(10,10))
plt.subplot(121), plt.imshow(sobel_150, cmap='gray') 
plt.title("Denoised Image (Threshold=150)"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(sobel_200, cmap='gray') 
plt.title("Denoised Image (Threshold=200)"), plt.xticks([]), plt.yticks([])
plt.show()
```

##### As shown in the results above, denoising an image before edge detection improves the precision of the detection. More specifically, the edge detections are much smoother especially on the facial features and contain much less noise than the original detections.
